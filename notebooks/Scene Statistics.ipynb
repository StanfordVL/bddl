```python
def scene_type(sn):
    if "_int" in sn or "_garden" in sn or "house_" in sn:
        return "house"
    if "office_" in sn:
        return "office"
    if "grocery_store" in sn:
        return "grocery store"
    if "restaurant" in sn:
        return "restaurant"
    if "hotel" in sn:
        return "hotel"
    if "school" in sn:
        return "school"
    if "hall_" in sn:
        return "generic hall"
    raise ValueError(sn)
```

```python
# Let's make the scene/room chart first
import collections
data = [(scene_type(scene.name), room.type.replace("_", " ")) for scene in Scene.all_objects() for room in scene.rooms if scene.name != "gates_bedroom"]
data = [{"source": k[0], "target": k[1], "value": v} for k, v in collections.Counter(data).items()]
```

```python
data
```

```python
scene_count_per_type = collections.Counter(scene_type(s.name) for s in Scene.all_objects() if s.name != "gates_bedroom")
room_types = {room.type.replace("_", " ") for room in Room.all_objects()}
total_value_per_room_type = {rt: sum(x["value"] for x in data if x["target"] == rt) for rt in room_types}
for d in data:
    source = d["source"]
    source_cnt = scene_count_per_type[source]
    source_str = f"{source.title()} ({source_cnt})"
    value = d["value"]
    target = d["target"]
    target_cnt = total_value_per_room_type[target]
    target_str = f"{target.title()} ({target_cnt})"
    # print(f'{source_str} [{value}] {target_str}')
    print(f'{target_str} [{value}] {source_str}')
```
